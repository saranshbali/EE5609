Given second degree equation after finding value of k=-3:

$2x^2+ xy -y^2 - 3x + 6y - 9 = 0$

Here, a=2, b = 1/2, c = -1, d=-3/2, e= 3,f=-9
 

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
V=np.array([[2,1/2],[1/2,-1]])
print("V =\n",V)
u=np.array([[-3/2],[3]])
print("u =\n",u)
f=-9
print("f =",f)
det=np.block([[V,u],[u.T,f]])
print("det =\n",det)
det_value = round(np.linalg.det(det))
print("Value of determinant is:",det_value)
```

Zero value of determinant implies that the above second degree equations represent a pair of straight lines.  

```python
x = np.linspace(-10,10,80)
y_1 = 2*x+3 # straight line 1
y_2 = -x+3 # straighe line 2
plt.plot(x, y_1, '-r', label='2x-y+3=0')
plt.title('Graph of 2y=3x+7')
plt.plot(x, y_2, '-b', label='x+y-3=0')
plt.title('Graph of two straight lines')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='best')
plt.grid()
plt.show()
```
