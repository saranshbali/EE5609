```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 140
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
fig =plt.figure()
ax = fig.add_subplot(111, projection = '3d')
x=np.array([2, -1, 1])
y=np.array([1, -3, -5])
z=np.array([3,-4 ,-4 ]) 
ax.scatter(x,y,z)
verts = [list(zip(x, y, z))]
plt.gca().add_collection3d(Poly3DCollection(verts,alpha =.5, facecolor='#800005'))
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
for x,y,z,i in zip(x,y,z,["A","B","C"]):
    ax.text(x,y,z,i)
```

```python
A=np.array([2, -1, 1])
B=np.array([1, -3, -5])
C=np.array([3,-4 ,-4 ])
#Now checking for whether the triangle is right angled using inner product and finding the right angle vertex. 
ip_1 = (A-C).transpose().dot(B-C) 
ip_2 = (A-B).transpose().dot(C-B)
ip_3 = (B-A).transpose().dot(C-A)
if(ip_1==0):
    print("Triangle ABC is right angled triangle, right angled at C.")
elif(ip_2==0):
    print("Triangle ABC is right angled triangle, right angled at B.")
elif(ip_3==0):
    print("Triangle ABC is right angled triangle, right angled at A.")
else:
    print("Triangle ABC is not a right angled triangle.")
```
